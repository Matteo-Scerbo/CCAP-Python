## Scikit-learn

Scikit-learn is a simple and efficient tools for data mining and data analysis in Python.
It is built on NumPy, SciPy, and matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
```

#### Dataset

```python
digits = datasets.load_digits()

X = digits.data
y = digits.target

print(X.shape)
print(y.shape)
print(np.unique(y))
```

```python
plt.figure(figsize=(15, 5))
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.imshow(X[i,:].reshape((8, 8)), cmap='gray')
    plt.title(y[i])
    plt.axis('off')
```

#### Data Preparation


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

#### Classification

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5) # definition of the class
```

Look at http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html for a documentation of the K-Neighbors classifier

```python
clf.fit(X_train, y_train) 
```

This learns a fitting between the input representation and the expected output

```python
idx = 8
y_hat = clf.predict(X_test[idx:idx+1])

plt.figure(figsize=(2, 2))
plt.imshow(X_test[idx].reshape((8, 8)), cmap='gray')
plt.axis('off')
plt.show()

print('True label:      %d' % y_test[idx])
print('Estimated label: %d' % y_hat)
```

```python
clf.score(X_test, y_test)
```
